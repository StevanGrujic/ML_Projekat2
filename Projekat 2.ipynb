```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="white")
sns.set(style="whitegrid", color_codes=True)

import warnings
```

```python
pd.set_option('display.max_columns', None)
df = pd.read_csv("data/marketing_campaign.csv", sep="\t")
df.head(100)
```

```python
df.groupby("Dt_Customer").median()
```

```python
kategoricki_atributi = ['Education','Marital_Status']
for a in kategoricki_atributi:
    fig, ax = plt.subplots();
    sns.countplot(data=df, y=a, ax=ax);
```

```python
df_corr = df.corr()
df_corr = df_corr.apply(lambda x: round(x, 2))
fig, ax = plt.subplots(figsize = (20,20));
sns.heatmap(data=df_corr, annot=True, ax=ax);
```
